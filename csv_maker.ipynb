```python
import pandas as pd
import numpy as np
```

```python
aerobic_df = pd.DataFrame()
people = []
for i in range(1, 10):
    people.append(f'f0{i}')
for i in range(10, 19):
    people.append(f'f{i}')
print(people)
```

```python
def find_stats(stat):
    # find max number of columns needed for output df
    max_cols = 0
    for person in people:
        sample_df = pd.read_csv(f'Wearable_Dataset/STRESS/{person}/{stat}.csv')
        if sample_df.shape[0] > max_cols:
            max_cols = sample_df.shape[0]
    # make df from dictionary
    output_dict = {}
    for person in people:
        # find data for that person
        values = pd.read_csv(f'Wearable_Dataset/STRESS/{person}/{stat}.csv').iloc[:, 0].tolist()
        values = values[1:] # cut off the hertz rate
        if len(values) < max_cols:
            values += [np.nan for i in range(max_cols - len(values))] # fill with missing values
        output_dict[person] = np.array(values)
    
    output = pd.DataFrame(output_dict).T
    return output
```

```python
aerobic_heartrate_df = find_stats('HR')
aerobic_heartrate_df
```
